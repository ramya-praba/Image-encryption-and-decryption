```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the BGR image
img = cv2.imread('/content/Kalpana_chawla.jpg')

# Split the image into its three channels
b, g, r = cv2.split(img)

# Set the histogram parameters
histSize = 256  # Number of bins
histRange = (0, 256)  # Range of possible pixel values

# Compute the histograms for each channel
hist_b = cv2.calcHist([b], [0], None, [histSize], histRange)
hist_g = cv2.calcHist([g], [0], None, [histSize], histRange)
hist_r = cv2.calcHist([r], [0], None, [histSize], histRange)

# Plot the histograms
plt.plot(hist_b, color='b')
plt.plot(hist_g, color='g')
plt.plot(hist_r, color='r')
plt.xlim([0, 256])
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.show()
# For encrypted image
img1 = cv2.imread('/content/final_encrypted_img_kalpana.png')
b1, g1, r1 = cv2.split(img1)

# Set the histogram parameters
histSize1 = 256  # Number of bins
histRange1 = (0, 256)  # Range of possible pixel values

# Compute the histograms for each channel
hist_b1 = cv2.calcHist([b1], [0], None, [histSize1], histRange1)
hist_g1 = cv2.calcHist([g1], [0], None, [histSize1], histRange1)
hist_r1 = cv2.calcHist([r1], [0], None, [histSize1], histRange1)

# Plot the histograms
plt.plot(hist_b1, color='b')
plt.plot(hist_g1, color='g')
plt.plot(hist_r1, color='r')
plt.xlim([0, 256])
plt.xlabel('Pixel value')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import cv2
def correlation_coefficient(image1, image2):
    # Convert images to float arrays
    image1 = np.array(image1, dtype=float)
    image2 = np.array(image2, dtype=float)

    # Subtract the mean from each image
    image1_mean = np.mean(image1)
    image2_mean = np.mean(image2)
    image1 -= image1_mean
    image2 -= image2_mean
    # Compute the correlation coefficient
    numerator = np.sum(image1 * image2)
    denominator = np.sqrt(np.sum(image1 ** 2) * np.sum(image2 ** 2))
    corr_coeff = numerator / denominator

    return corr_coeff

image1=cv2.imread('/content/Kalpana_chawla.jpg')
image2=cv2.imread('/content/final_encrypted_img_kalpana.png')
x=correlation_coefficient(image1,image2)
print(x)
```

```python
#HCC
import cv2
import numpy as np

# Load the image
img = cv2.imread('/content/final_encrypted_img_kalpana.png')

# Split the image into its color channels
b, g, r = cv2.split(img)

# Compute the HCC for each color channel
hcc_b = np.corrcoef(b[:, :-1], b[:, 1:])[0, 1]
hcc_g = np.corrcoef(g[:, :-1], g[:, 1:])[0, 1]
hcc_r = np.corrcoef(r[:, :-1], r[:, 1:])[0, 1]
avg = (hcc_b+hcc_g+hcc_r)/3
# Print the results
print(f'HCC (blue channel): {hcc_b}')
print(f'HCC (green channel): {hcc_g}')
print(f'HCC (red channel): {hcc_r}')
print(avg)
```
